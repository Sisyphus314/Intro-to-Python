## Question 2

Write a function called "get_corr_pairs" which takes one argument:

* df, which is a pandas data frame

and returns:

* corr_pairs, a dictionary where keys are names of columns of df corresponding to numerical features, and values are arrays of names of columns whose correlation coefficient with the key has magnitude 0.3 or greater. 

You can use your function from question 1 to get the correlation values.

````python
#### play with code here #####
from pprint import pprint
def get_corr_pairs(dataframe):
    corr_max = get_corrs(dataframe)
    col_list = list(corr_max.columns)
    pair_dict = {}
    pair_dict = {col: None for col in col_list}
    for col in col_list:
        pair_list = []
        for i in range(0, len(col_list)):
            if abs(corr_max.loc[col][i]) > 0.3 and abs(corr_max.loc[col][i]) != 1:
                pair_list.append(col_list[i])
        pair_dict.update({col:pair_list})
    return pprint(pair_dict)
````

## Question 3

Write a function called "sample_cereal" which takes two arguments:
    
* df, which is a pandas data frame
* kind, which is a string that can take value 'up' or 'down'

and returns:

* a pandas data frame with balanced target class 'ratingID', using up sampling if kind='up' and downsampling if kind='down'.

````python
#### play with code here #####

def sample_cereal(dataframe, kind):
    if kind == "up":
        count = dataframe.apply(lambda x: True if x['ratingID'] == 1 else False, axis=1)
        count1 = len(count[count == True].index)
        count0 = len(count[count == False].index)
        if count1 > count0:
            #major group = rating 1
            #minor group = rating 0
            #we need to up minor group
            #we will do random duplicating from minor group
            diff = count1 - count0
            #diff = how many duplicates we need
            df0 = dataframe.loc[dataframe["ratingID"] == 0]
            random0 = df0.sample(n = diff, random_state = 1)
            #append random0 to dataframe
            return dataframe.append(random0)
        elif count1 < count0:
            #major = rating0
            # up the minor rating 1 group
            diff = count0 - count1
            #find minor group
            df1 = dataframe.loc[dataframe["ratingID"] == 1]
            #random sample minor group
            random1 = df1.sample(n = diff, random_state = 1)
            #append random1 to dataframe
            return dataframe.append(random1)
        elif count1 == count0:
            print ("The dataset is in balance")
    elif kind == "down":
        # down sampling
        count = dataframe.apply(lambda x: True if x['ratingID'] == 1 else False, axis=1)
        count1 = len(count[count == True].index)
        count0 = len(count[count == False].index)
        if count1 > count0:
            #major = rating1
            #drop random rows in rating1
            diff = count1 - count0
            df1 = dataframe.loc[dataframe["ratingID"] == 1]
            random1 = df1.sample(n = diff, random_state = 1)
            random1_list = list(random1.index.values)
            return dataframe.drop(random1_list)
            
        elif count1 < count0:
            #major = rating0
            #drop random rows in rating0
            diff = count0 - count1
            df0 = dataframe.loc[dataframe["ratingID"] == 0]
            random0 = df0.sample(n = diff, random_state = 1)
            random0_list = list(random0.index.values)
            return dataframe.drop(random0_list)
        elif count1 == count0:
            print ("The dataset is in balance.")
````

## Question 4

Write a function called "find_H" which takes two arguments:

* df, which is a pandas data frame
* cname, which is the name of the target column (that should correspond to a categorical variable)

and returns:

* H, the entropy in the column cname (use logarithm base 2)

````python
#### play with code here #####
import math
def find_H(dataframe, cname):
    count_bool = dataframe.apply(lambda x: True if x[cname] == 1 else False, axis=1)
    count0 = len(count_bool[count_bool == False].index)
    count1 = len(count_bool[count_bool == True].index)
    x, y = dataframe.shape
    if count0 == 0 or count1 == 0:
        return 0
    else:
        p0 = count0/x
        p1 = count1/x
        ent = -p0*math.log2(p0)-p1*math.log2(p1)
    return ent
````

## Question 5

We talked about using gini score for split in class, but we will focus on using entropy in this practice. 
What want to know how much information can be gained by splitting on one specific feature. 

Formally, given resp. variable $Y$, dataset $D$ and feature $X$, information gain of $Y$ from $X$ is defined as 

$ IG(Y, X) = H(Y) - H(Y|X) $

where $H(Y|X)$ is defined as

$H(Y|X) = \sum_{l \in levels(X)} P(l) \cdot H(l). $

See [here](https://medium.com/@rishabhjain_22692/decision-trees-it-begins-here-93ff54ef134) for an illustration.

Write a function called "info_gain" which takes four arguments:

* df, which is a pandas data frame
* cname, which is the name of the target column (that should correspond to a categorical variable)
* csplit, which is the name of a numeric column in df (aka, feature)
* threshold, which is a numeric value (aka, split cutoff)

and returns:

* info_gain, the information gain you get in column cname by splitting the dataset on the threshold value in column csplit. 

````python
#### play with code here #####
def info_gain(dataframe, cname, csplit, threshold):
    threshold = float(threshold)
    x, y = dataframe.shape
    
    master_e = find_H(dataframe, cname)
    
    threshold_up = dataframe[csplit] >= threshold
    up_e = find_H(dataframe[threshold_up], cname)*len(dataframe[threshold_up])/x
    
    threshold_down = dataframe[csplit] < threshold
    down_e = find_H(dataframe[threshold_down], cname)*len(dataframe[threshold_down])/x
    
    gain_info = master_e - up_e - down_e
    return gain_info
````
