```python
def get_manager(employee):
    emp = employee[['EmployeeID', 'FirstName', 'MiddleName', 'LastName','ManagerID']]
    manager = employee[['EmployeeID', 'FirstName', 'LastName']]
    manager.columns = ['EmployeeID_man', 'ManagerFirstName', 'ManagerLastName']
    emp_man = pd.merge(emp, manager, 'left', 
                      left_on = ['ManagerID'],
                      right_on = ['EmployeeID_man'])
    
    emp_man = emp_man.drop(['ManagerID', 'EmployeeID_man'], axis=1)
    
    return emp_man
```

Write a function called get_spend_by_order that takes as its two arguments the Pandas DataFrames "Orders" and "Customers", and returns a DataFrame with the following columns: "FirstName", "LastName", "Item", "TotalSpent", listing all customer names, their purchased items, and the total amount spent on that item (remember that the "Price" listed in "Orders" is the price per item).

Hint: Consider which ID column to merge on. 

```python
Customers = pd.read_excel("~/Desktop/DataProg/Customers.xls")
Orders = pd.read_excel("~/Desktop/DataProg/ItemsOrdered.xls")
def get_spend_by_order(Orders, Customers):
    customer = pd.read_excel("~/Desktop/DataProg/Customers.xls")
    order = pd.read_excel("~/Desktop/DataProg/ItemsOrdered.xls")
    cus = customer[["CustomerID", "FirstName", "LastName"]]
    cus_order = order[["CustomerID", "Item", "Quantity", "Price"]]
    cus_order["TotalSpent"] = cus_order["Quantity"]*cus_order["Price"]
    money = pd.merge(cus, cus_order, on = "CustomerID")
    money = money.drop(["CustomerID", "Quantity", "Price"], axis = 1)
    return money
```

Write a function called get_order_location that takes three arguments: "Orders", "Customers", and "Territory", and returns a DataFrame containing the following columns: "CustomerID", "Name", and "TotalItems", that gives, for each order, the CustomerID, the name of the territory where the order was placed, and the total number of items ordered (yes, 2 ski poles counts as 2 items). 

Hint: Consider groupby then sum the quantities. 

```python
Territory = pd.read_excel("~/Desktop/DataProg/SalesTerritory.xls")
Customers = pd.read_excel("~/Desktop/DataProg/Customers.xls")
Orders = pd.read_excel("~/Desktop/DataProg/ItemsOrdered.xls")
```

```python
def get_order_loction(Orders, Customers, Territory):
    order = Orders[["CustomerID", "Quantity"]]
    cus = Customers[["CustomerID", "SalesTerritoryID"]]
    cus = cus.rename(columns = {"SalesTerritoryID": "TerritoryID"})
    terry = Territory[["TerritoryID", "Name"]]
    cus_terry = pd.merge(cus, terry, on = "TerritoryID")
    # merge customer with territory on territory id
    order_cus_terry = pd.merge(order, cus_terry, on = "CustomerID")
    order_loc = order_cus_terry.groupby(["CustomerID", "Name"])["Quantity"].sum()
    orderloc = order_loc.to_frame().reset_index()
    orderloc = orderloc.rename(columns = { 0: "TotalItems"})
    return orderloc
```

```python
get_order_loction(Orders, Customers, Territory)
```

Write a function called employee_info that takes one argument: "Employees", and returns a DataFrame containing the following columns: JobTitle, NumberOfEmployees, and MeanVacationHours, containing all job titles, the number of employees with that job title, and the mean number of vacation hours for employees with that job title.

Hint: Consider groupby then count. 

```python
Employees = pd.read_excel("~/Desktop/DataProg/Employees.xls")
```

```python
def employee_info(Employees):
    employee = Employees[["JobTitle", "EmployeeID"]]
    job_num = employee.groupby(["JobTitle"]).count()
    job_num = job_num.rename(columns = { "EmployeeID": "NumberOfEmployees"})
    job_num.reset_index()
    hours = Employees[["JobTitle", "VacationHours"]]
    hour = hours.groupby(["JobTitle"])["VacationHours"].sum()
    hour = hour.to_frame().reset_index()
    emp_hour = pd.merge(job_num, hour, on = "JobTitle")
    emp_hour["MeanVacationHours"] = emp_hour["VacationHours"] / emp_hour["NumberOfEmployees"]
    emp_hour = emp_hour.drop(["VacationHours"], axis=1)
    return emp_hour
```

```python
employee_info(Employees)
```
